```python
# The following uses Python 3.7, because that's where I have obspy and cartopy installed


import obspy as ob
import numpy as np
from obspy.clients.fdsn import Client
import matplotlib.pyplot as plt
from obspy.signal.util import util_geo_km
import re

# Full Map
import cartopy
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import cartopy.feature as cfeature
import cartopy.io.img_tiles as cimgt
import matplotlib.ticker as mticker



%matplotlib inline
```

```python
iris_client = Client("IRIS")

# https://earthquake.usgs.gov/research/parkfield/seismic.php


# Info on the array:
# https://ds.iris.edu/mda/XN/?starttime=2000-01-01T00:00:00&endtime=2002-12-31T23:59:59

# An example station, BHZ
# https://ds.iris.edu/mda/XN/AHAB/--/BHZ/?starttime=2001-06-24T00:00:00&endtime=2002-06-04T23:03:14

st = ob.UTCDateTime("2001-10-24T00:00:00")
et = st+60
bulk_request = [("XN", "*", "--", "BHZ", st, et)]
inventory = iris_client.get_stations_bulk(bulk_request, level='channel') # or 'response' for more
print(inventory)
```

```python
lats = [ sta.latitude for sta in inventory[0].stations]
lons = [ sta.longitude for sta in inventory[0].stations]
elevs = [ sta.elevation for sta in inventory[0].stations]
names = [ sta.code for sta in inventory[0].stations]

# For Andreas' code, station locations must be defined in meters relative to a reference point.
ref_lat = np.mean(lats)
ref_lon = np.mean(lons)

xx=np.zeros(np.size(inventory[0]),)
yy=np.zeros(np.size(inventory[0]),)
#distance = ob.geodetics.calc_vincenty_inverse(ref_lat, ref_lon, lats[0], lons[0])
for i in range(np.size(inventory[0])):
    xx[i],yy[i] = util_geo_km(ref_lon, ref_lat, lons[i], lats[i])

print(lats, lons, names)
print(xx,yy)
```

```python
for i in range(np.size(xx)):
    print("{0:e}".format(xx[i]*1000))
```

```python
# Quick visual check
plt.plot(lons,lats,'rv')
plt.show()

plt.plot(xx,yy,'rv')
plt.xlabel('Distance (km)')
plt.ylabel('Distance (km)')
plt.show()
```

```python
# Full Map

def scale_bar(ax, length=None, location=(0.5, 0.05), linewidth=3):
    """
    ax is the axes to draw the scalebar on.
    length is the length of the scalebar in km.
    location is center of the scalebar in axis coordinates.
    (ie. 0.5 is the middle of the plot)
    linewidth is the thickness of the scalebar.
    """
    #Get the limits of the axis in lat long
    llx0, llx1, lly0, lly1 = ax.get_extent(ccrs.PlateCarree())
    #Make tmc horizontally centred on the middle of the map,
    #vertically at scale bar location
    sbllx = (llx1 + llx0) / 2
    sblly = lly0 + (lly1 - lly0) * location[1]
    tmc = ccrs.TransverseMercator(sbllx, sblly)
    #Get the extent of the plotted area in coordinates in metres
    x0, x1, y0, y1 = ax.get_extent(tmc)
    #Turn the specified scalebar location into coordinates in metres
    sbx = x0 + (x1 - x0) * location[0]
    sby = y0 + (y1 - y0) * location[1]

    #Calculate a scale bar length if none has been given
    #(Theres probably a more pythonic way of rounding the number but this works)
    if not length: 
        length = (x1 - x0) / 5000 #in km
        ndim = int(np.floor(np.log10(length))) #number of digits in number
        length = round(length, -ndim) #round to 1sf
        #Returns numbers starting with the list
        def scale_number(x):
            if str(x)[0] in ['1', '2', '5']: return int(x)        
            else: return scale_number(x - 10 ** ndim)
        length = scale_number(length) 

    #Generate the x coordinate for the ends of the scalebar
    bar_xs = [sbx - length * 500, sbx + length * 500]
    #Plot the scalebar
    ax.plot(bar_xs, [sby, sby], transform=tmc, color='k', linewidth=linewidth)
    #Plot the scalebar label
    ax.text(sbx, sby, str(length) + ' km', transform=tmc,
            horizontalalignment='center', verticalalignment='bottom')



fig, ax = plt.subplots(1, figsize=(20,10))
proj = ccrs.PlateCarree()                       # the projection you want to use, many are supported, 
                                                # this one is the standard for geographic coordinates
 
llcrnrlon, llcrnrlat = -120.6,35.93                      # lower left corner longitude/latitude
urcrnrlon, urcrnrlat = -120.46,36.04                   # upper right corner longitude/latitude


ax = plt.axes(projection=proj)                  # sets the projection
ax.set_extent([llcrnrlon, urcrnrlon, llcrnrlat, urcrnrlat])


gl = ax.gridlines(crs=ccrs.PlateCarree()        
                , draw_labels=True                 # label coordinates
                , linewidth=1                   # line properties can be changed similar to matplotlib
                , color='gray'
                , alpha=0.5
                , linestyle='--')

gl.xlabels_top = gl.ylabels_right = False                          # where labels shouldn't be
step = 0.05
gl.xlocator = mticker.FixedLocator(np.arange(llcrnrlon-step,urcrnrlon+step,step))           # sets longitude tick marks
gl.ylocator = mticker.FixedLocator(np.arange(llcrnrlat-step,urcrnrlat+step,step))  
gl.xformatter = LONGITUDE_FORMATTER                               # set cartopy formatting
gl.yformatter = LATITUDE_FORMATTER
gl.xlabel_style = {'size': 12, 'color': 'black'}
gl.ylabel_style = {'size': 12, 'color': 'black'}
print(np.arange(llcrnrlon,urcrnrlon,.005))

ax.scatter(lons,lats               # plots your data points on the axes
        , marker='v'                                 # symbol for data point
        , c='red'                                    # symbol facecolor
        , edgecolor='black'                          # symbol edgecolor
        , alpha=0.9                                  # opacity
        , s = 75                                     # size
        , transform=ccrs.PlateCarree())              # projection to plot data in

for i in range(len(names)):
    name=re.split('-',names[i])
    ax.text(lons[i]+.0005,lats[i], name[0], va="center", ha="left", family="monospace", weight="bold").set_clip_on(True)

scale_bar(ax, 1)
    
plt.title('Parkfield Array')
plt.savefig('array.pdf', bbox_inches='tight', format='pdf')

plt.show()



```

```python
# Format for Andreas' code
file = open("INPUT/receivers_all.txt","w")
file.write("#- Number of receivers. -------------------------\n")
file.write("Nreceivers={0}\n".format(np.size(xx)))
file.write("\n")
file.write("#- Receiver positions. --------------------------\n")
for i in range(np.size(xx)):
    file.write("x{0}={1:e}\n".format(i, xx[i]*1000))
    file.write("y{0}={1:e}\n".format(i, yy[i]*1000))
    file.write("\n")
file.close()

# Use a subset of stations for now

names_use = ['SAGE','AHAB','PINE','BECH','KARL','PRIS','POND','ALEX','STGI','GLEN','DBLT']
indices_use = []
for n in names_use:
    indices_use.append(names.index(n))

file = open("INPUT/receivers.txt","w")
file.write("#- Number of receivers. -------------------------\n")
file.write("Nreceivers={0}\n".format(np.size(indices_use)))
file.write("\n")
file.write("#- Receiver positions. --------------------------\n")
for i in range(np.size(indices_use)):
    file.write("x{0}={1:e}\n".format(i, xx[indices_use[i]]*1000))
    file.write("y{0}={1:e}\n".format(i, yy[indices_use[i]]*1000))
    file.write("\n")
file.close()


### Should look like this
# #- Number of receivers. -------------------------
# Nreceivers=15

# #- Receiver positions. --------------------------
# x0=-2.0e6
# y0=0.0e6

# x1=2.0e6
# y1=0.0e6
```
