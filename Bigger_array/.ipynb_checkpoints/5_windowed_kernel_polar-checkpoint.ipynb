```python
# Quick single pair test


i=7
j=10

counter = 0
for i1 in range(0,p.Nreceivers):
    for i2 in range(0,p.Nreceivers):
        if(i1==i and i2==j):            
            final_counter = counter
        counter += 1 # 
counter = final_counter

# Distance vector (in km)
distance_x = xx[i]-xx[j]
distance_y = yy[i]-yy[j]


# search for window start and end - checking edges is probably enough
t_window_min = np.max(t)
t_window_max = np.min(t)
# for isx in np.arange(min_sx, max_sx, dux):
#     for isy in np.arange(min_sy, max_sy, duy):
for theta in thetas:
    for radius in radii:
        isx = radius*np.cos(theta)
        isy = radius*np.sin(theta)
        time_effective = -isx*distance_x - isy*distance_y

        if time_effective < t_window_min:
            t_window_min = time_effective
        if time_effective > t_window_max:
            t_window_max = time_effective

# Now we have window. Zero out other times before feeding into source_kernel()?
i_start =  (np.abs(t - t_window_min)).argmin()
i_end   =  (np.abs(t - t_window_max)).argmin()
correlations_windowed = np.zeros(np.shape(correlations[:,counter]))
correlations_windowed[i_start:i_end] = correlations[i_start:i_end,counter]
            
    
# Plot stuff
plt.figure(figsize=(10,5))
plt.plot(t,correlations[:,counter],'k')
plt.plot(t,correlations_windowed,'b')

ax = plt.gca()
[y_min,y_max]=ax.get_ylim()
# plt.plot([t[index_time],t[index_time]], [y_min,y_max], 'r')
plt.title("{0} - {1}, index {2}".format(i,j,counter))
plt.show()            

            

# Plot kernels
x,y,K_temp = source_kernel(correlations[:,counter], t, rec0=i, rec1=j, measurement='amp', effective=0, plot=1)
x,y,K_temp = source_kernel(correlations_windowed, t, rec0=i, rec1=j, measurement='amp', effective=0, plot=1)
```

```python
K_all = np.zeros([p.Nreceivers**2-p.Nreceivers,np.shape(x)[0],np.shape(x)[1]])


# Loop through station pairs
counter = 0
for i1 in range(0,p.Nreceivers):
    for i2 in range(0,p.Nreceivers):
        if(i1==i2):
            counter += 1 # Ignore autocorrelations, but up the counter
            continue
        print(i1,i2)
        
        
        # Distance vector (in km)
        distance_x = xx[i1]-xx[i2]
        distance_y = yy[i1]-yy[i2]


        # search for window start and end - checking edges is probably enough
        t_window_min = np.max(t)
        t_window_max = np.min(t)
        for theta in thetas:
            for radius in radii:
                isx = radius*np.cos(theta)
                isy = radius*np.sin(theta)
                time_effective = -isx*distance_x - isy*distance_y

                if time_effective < t_window_min:
                    t_window_min = time_effective
                if time_effective > t_window_max:
                    t_window_max = time_effective

        # Now we have window. Zero out other times before feeding into source_kernel()?
        i_start =  (np.abs(t - t_window_min)).argmin()
        i_end   =  (np.abs(t - t_window_max)).argmin()
        correlations_windowed = np.zeros(np.shape(correlations[:,counter]))
        correlations_windowed[i_start:i_end] = correlations[i_start:i_end,counter]
            
        x,y,K_all[counter,:,:] = source_kernel(correlations_windowed, t, rec0=i1, rec1=i2, measurement='amp', effective=0, plot=0)
        
        counter += 1
        
K_mean = np.mean(K_all,0)
K_sum = np.sum(K_all,0)


# We know where the source actually is:
x0=0.75*p.xmax
y0=0.75*p.ymax


np.save('5_Ksum_windowed_square', K_sum)
        
#                         index_time = (np.abs(t - time_effective)).argmin()
#                         P[ix,iy] += correlations[index_time,counter]
```

```python
mpl.rc('font', **{'size':20})

def plot_K(K,title="MFP",save=0,savename='none',cmax=0):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_axes([0.1,0.1,0.6,0.6])
    cmap = plt.get_cmap('RdBu')
    i = plt.pcolor((x-p.dx/2)/1000,(y-p.dy/2)/1000,K,cmap='RdBu',rasterized=True)#,vmin=-4,vmax=4)
    if(cmax==0):
        cmax=np.max(np.abs(K))*0.15
    plt.clim(-cmax,cmax)
    plt.xlabel('Distance [km]')
    plt.ylabel('Distance [km]')
    plt.plot(x0/1000.0,y0/1000.0,'r*')
    plt.plot(xx,yy,'k^')
    plt.title(title)
    plt.axis('equal')
    ax.tick_params(top=True,right=True)
    colorbar_ax = fig.add_axes([0.75, 0.1, 0.03, 0.6])
    fig.colorbar(i, cax=colorbar_ax)
    if(save==1):
        plt.savefig(savename, bbox_inches='tight')
    plt.show()

plot_K(K_all[1,:,:],title="Source kernel [unit of measurment / m^2]",save=1,savename="4_MFP_by_kernels.pdf")

plot_K(K_sum,title="Source kernel [unit of measurment / m^2]",save=1,savename="4_MFP_by_kernels.pdf")
```
